```python
import pandas as pd
import matplotlib.pyplot as plt

dim_to_plot = "Pitch"
# file_path = "./carla_data/imu_1729669097.117631.csv"
file_path = "carla_data/output_1729673237.8122728.csv"
df = pd.read_csv(file_path)
```

```python
# Plot timestamp (normalized) versus angular_velocity_z
plt.figure(figsize=(10, 6))
plt.plot(
    df["Timestamp"],
    df[dim_to_plot],
    label="",
    color="b",
)

plt.legend()
plt.grid(True)
plt.show()
```
